```python
def data_process(df):
    df = df.pct_change()
    df = df.tail(-1)
    df = df + 1
    df = df.cumprod()
    df = df - 1
    df = df.iloc[-1,:]
    return df
```

```python
# RMSE
def RMSE(duration, index_return, portfolio_return):
    RMSE = 0
    for i in range(duration):
        RMSE += (index_return[i] - portfolio_return[i]) ** 2
    return math.sqrt(RMSE/duration)
```
